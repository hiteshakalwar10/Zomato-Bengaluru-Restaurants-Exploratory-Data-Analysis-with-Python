```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
#TOP CUISINES
cuisines_series = df["cuisines"].str.split(", ")

all_cuisines = []

for cuisine_list in cuisines_series:
    all_cuisines.extend(cuisine_list)

cuisine_counts = pd.Series(all_cuisines).value_counts().head(10)

print(cuisine_counts)

plt.figure(figsize=(8,5))
plt.bar(cuisine_counts.index, cuisine_counts.values)

plt.title("Top 10 Cuisines")
plt.xlabel("Cuisines")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)
plt.show()
```

```python
#RATING DISTRIBUTION
plt.figure(figsize=(8,5))

sns.histplot(df["rate"], bins=10)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=90)

plt.show()
```

```python
#COST ANALYSIS
plt.figure(figsize=(8,5))

sns.histplot(df["approx_cost(for two people)"], bins=10)

plt.title("Cost Distribution")
plt.xlabel("Cost for Two")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
#ONLINE vs OFFLINE
online_counts = df["online_order"].value_counts()

print(online_counts)
plt.figure(figsize=(5,4))

plt.bar(online_counts.index, online_counts.values)

plt.title("Online vs Offline Orders")
plt.xlabel("Order Type")
plt.ylabel("Count")

plt.show()
```

```python
#area wise analysis
top_locations = df["location"].value_counts().head(10)

print(top_locations)
plt.figure(figsize=(10,5))

plt.bar(top_locations.index, top_locations.values)

plt.title("Top 10 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")

plt.xticks(rotation=45)
plt.show()
```
